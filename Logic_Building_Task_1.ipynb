 1. User Login Check

```python
#Given Credentials
stored_user = "admin"
stored_pass = "1234"

# Taking input from user
print("--- Login System ---")
username = input("Enter Username: ")
password = input("Enter Password: ")

# Checking credentials
if username == stored_user and password == stored_pass:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters for passed and failed students
pass_count = 0
fail_count = 0

# Iterate through each mark to determine pass/fail status
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final results
print(f"Total students: {len(marks)}")
print(f"Number of students who passed: {pass_count}")
print(f"Number of students who failed: {fail_count}")
```

3. Simple Data Cleaner

```python
# List of names
names = [" Alice ", "bob", " CHARLIE "]

# List to store cleaned names
cleaned_names = []

# Process to remove extra spaces and convert to lowercase
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print the list of cleaned names
print("Original names:", names)
print("Cleaned names:", cleaned_names)
```

4. Message Length Analyzer

```python
# List of messages to analyze
messages = ["Hi", "Welcome to the platform", "OK", "A long message that exceeds ten characters"]

print("message length analysis:")
for msg in messages:
    msg_length = len(msg)
    print(f"- Message: '{msg}', Length: {msg_length}", end="")
    # Flag messages longer than 10 characters
    if msg_length > 10:
        print(" (FLAG: Longer than 10 characters)")
    else:
        print()
```

5. Error Message Detector

```python
# List of log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize counter
error_count = 0

# Iterations
for entry in logs:
    if entry == "ERROR":
        error_count += 1

# Print the total number of error messages detected
print(f"System Logs: {logs}")
print(f"Total 'ERROR' entries found: {error_count}")
```
