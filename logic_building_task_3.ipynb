Problem 1: Employee Performance Bonus Eligibility

```python
# Dictionary of employee names and their scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# First, we need to find out what the highest score is
top_score = 0
for name, score in employees.items():
    if score > top_score:
        top_score = score

# Now find everyone who has that top score
top_performers = []
for name, score in employees.items():
    if score == top_score:
        top_performers.append(name)

# formatting the output to look like the example
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {top_score})")
```

Problem 2: Search Query Keyword Analysis

```python
# The search query string
query = "Buy mobile phone buy phone online"

# dealing with case sensitivity and splitting into words
# usually good to strip out common punctuation just in case, though this input is clean
processed_query = query.lower().replace('.', '').replace(',', '')
words = processed_query.split()

# dictionary to keep track of counts
word_counts = {}

# counting how many times each word appears
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# filtering to keep only words that show up more than once
duplicates = {}
for word, count in word_counts.items():
    if count > 1:
        duplicates[word] = count

print(duplicates)
```

Problem 3: Sensor Data Validation

```python
# Raw sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# list to store the valid tuples
valid_readings = []

# Using enumerate to get both the index (hour) and the value
for hour, reading in enumerate(sensor_readings):
    # checking if the reading is even
    if reading % 2 == 0:
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem 4: Email Domain Usage Analysis

```python
# List of employee emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}
total_emails = len(emails)

# Looping through to extract domains
for email in emails:
    # splitting by '@' gives us the name at index 0 and domain at index 1
    domain = email.split('@')[1]

    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1

# Calculating percentage and printing
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    # converting to int to match the expected output format (e.g., 60%)
    print(f"{domain}: {int(percentage)}%")
```

Problem 5: Sales Spike Detection

```python
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average sales first
total_sales = sum(sales)
average_sales = total_sales / len(sales)

# calculating the threshold (30% above average)
threshold = average_sales * 1.30

# Check each day to see if it spiked
# usually days are 1-indexed for humans, so i'll add 1 to the index
for i, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {i + 1}: {sale}")
```

Problem 6: Duplicate User ID Detection

```python
# List of user IDs with some potential duplicates
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to count frequencies
id_counts = {}

for user in user_ids:
    if user in id_counts:
        id_counts[user] += 1
    else:
        id_counts[user] = 1

# Displaying only the duplicates
for user, count in id_counts.items():
    if count > 1:
        print(f"{user} {count} times")
```
