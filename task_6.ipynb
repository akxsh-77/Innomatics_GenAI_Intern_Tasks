Problem 1: Smart Parking Lot Management System

```python
def manage_parking_lot(capacity, logs):
    # Initialize the count of parked vehicles
    parked_count = 0

    for log in logs:
        if log == "IN":
            # Only increment if there is space, otherwise we'd go over capacity
            if parked_count < capacity:
                parked_count += 1
        elif log == "OUT":
            # Only decrement if there are vehicles actually in the lot
            if parked_count > 0:
                parked_count -= 1

    # Determine the status based on occupancy
    if parked_count >= capacity:
        status = "Full"
    elif parked_count >= capacity * 0.9:
        status = "Almost Full"
    else:
        status = "Available"

    return parked_count, status

# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Execution
currently_parked, parking_status = manage_parking_lot(capacity, vehicle_logs)

print(f"Currently Parked Vehicles: {currently_parked}")
print(f"Parking Status: {parking_status}")
```

Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance_km, traffic_level, weather):
    # Base time: assuming 5 minutes per kilometer
    base_time = distance_km * 5

    # Traffic adjustments
    traffic_delay = 0
    if traffic_level.lower() == "high":
        traffic_delay = 10
    elif traffic_level.lower() == "medium":
        traffic_delay = 5

    # Weather adjustments
    weather_delay = 0
    if weather.lower() == "rainy":
        weather_delay = 5
    elif weather.lower() == "stormy":
        weather_delay = 15

    # Total ETA
    total_time = base_time + traffic_delay + weather_delay
    return total_time

# Sample Input
dist = 8
traffic = "High"
weather_cond = "Rainy"

# Execution
eta = estimate_delivery_time(dist, traffic, weather_cond)
print(f"Estimated Delivery Time: {eta} minutes")
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def analyze_theatre_occupancy(total_seats, booked_seats_list):
    # Total booked is the length of the list of seat entries
    num_booked = len(booked_seats_list)

    # Calculate occupancy percentage
    occupancy_pct = (num_booked / total_seats) * 100

    # Determine status
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return occupancy_pct, status

# Sample Input
total = 200
# Example of 150 entries as per prompt requirements
booked = [1] * 150

# Execution
occ, show_status = analyze_theatre_occupancy(total, booked)
print(f"Occupancy: {int(occ)}%")
print(f"Show Status: {show_status}")
```

Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        return 0, "No Data"

    # Calculate the average CPU usage
    avg_load = sum(cpu_readings) / len(cpu_readings)

    # Logic for status classification
    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return avg_load, status

# Sample Input
readings = [45, 60, 70, 85, 90]

# Execution
average, server_status = classify_server_load(readings)
print(f"Average CPU Load: {int(average)}%")
print(f"Server Status: {server_status}")
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources(usage_dict):
    overused = []

    # Check each resource in the dictionary
    for resource, hours in usage_dict.items():
        if hours > 5: # Threshold for overuse
            overused.append(resource)

    # Determine if an alert is needed
    energy_alert = "Yes" if len(overused) > 0 else "No"

    return ", ".join(overused), energy_alert

# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Execution
overused_items, alert = monitor_resources(resources)
print(f"Overused Resources: {overused_items}")
print(f"Energy Alert: {alert}")
```

Problem 6: Online Event Registration Capacity Controller

```python
def control_event_registration(capacity, total_registrations):
    # If registrations are within capacity
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlist = 0
        status = "Open" if total_registrations < capacity else "Full"
    else:
        # If registrations exceed capacity
        confirmed = capacity
        waitlist = total_registrations - capacity
        status = "Closed"

    return confirmed, waitlist, status

# Sample Input
event_cap = 100
regs = 105

# Execution
conf, wait, reg_status = control_event_registration(event_cap, regs)
print(f"Confirmed Registrations: {conf}")
print(f"Waitlisted Users: {wait}")
print(f"Registration Status: {reg_status}")
```
