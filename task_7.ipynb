Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_action(self, action):
        # Update the status based on the action provided
        if action.upper() == "ON":
            self.status = "ON"
        else:
            self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Testing with sample input
light_name = "Bedroom Light"
action = "ON"

my_light = SmartLight(light_name)
my_light.set_action(action)
my_light.display_status()
```

Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def show_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Testing with sample input
emp = Employee("Rahul", "EMP102", "IT")
emp.show_id_card()
```

Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Testing with sample input
person = Contact("Anita", "9876543210")
person.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        # Formatting with the Rupee symbol as per the expected output
        print(f"Price: ₹{self.price}")

# Testing with sample input
item = Product("Headphones", 2499)
item.print_price_tag()
```

Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def show_details(self):
        print(f"Movie: {self.name}")
        # Displaying rating with the /5 format
        print(f"Rating: {self.rating}/5")

# Testing with sample input
film = Movie("Inception", 4.8)
film.show_details()
```

Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

# Testing with sample input
order = Delivery("Suman", "Hyderabad")
order.print_details()
```
