Problem 1: Unique Words in a Sentence

```python
# taking the input sentence
sentence = "python is easy and python is powerful"

# splitting the sentence into individual words
words_list = sentence.split()

# using a set to get unique words because sets don't allow duplicates
unique_words = set(words_list)

# printing the results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
# creating the dictionary with employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# variables to store the max salary and the person's name
max_salary = 0
top_employee = ""

# looping through the dictionary to find the highest earner
for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        top_employee = name

print(f"Highest Salary: {top_employee} - {max_salary}")
```

Problem 3: Find Maximum and Minimum Values

```python
# list of numbers
numbers = [45, 22, 89, 10, 66]

# initializing max and min with the first item in the list
max_val = numbers[0]
min_val = numbers[0]

# iterating through the list to compare values
for num in numbers:
    if num > max_val:
        max_val = num # found a new max
    if num < min_val:
        min_val = num # found a new min

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

Problem 4: Count Products Above a Price Threshold

```python
# list of product prices
prices = [450, 1200, 899, 1500, 300]

# counter for products above 1000
count = 0

# checking each price
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
# attendance record list
attendance = ["P", "P", "A", "P", "P"]

# calculating total days and present days
total_days = len(attendance)
days_present = attendance.count("P")

# finding the percentage
percentage = (days_present / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency

```python
input_str = "python"
char_counts = {}

# looping through each character in the string
for char in input_str:
    # if char is already in dict, add 1, else set it to 1
    if char in char_counts:
        char_counts[char] += 1
    else:
        char_counts[char] = 1

print(char_counts)
```

Problem 10: Calculate Average Marks

```python
marks = [85, 90, 78, 92, 88] # example marks list

# finding sum and length
total_marks = sum(marks)
count = len(marks)

# calculating average
average = total_marks / count

print("Average Marks:", average)
```
