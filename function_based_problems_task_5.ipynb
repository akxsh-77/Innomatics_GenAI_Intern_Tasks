Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(total_likes):
    # Analyzing if the post is viral based on the 1000 likes threshold
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Printing the result in the specific format required
    print(f"Total Likes: {total_likes} Post Status: {status}")

# Testing the function with the example value
analyze_engagement(1050)
```

Problem 2: Healthcare – Medicine Stock Alert System

```python
def monitor_stock(stock_level):
    # Checking if we are running low on medicine (less than 10 items)
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    # Displaying the stock count and the alert status
    print(f"Medicine Stock: {stock_level} Status: {status}")

# Running the check with a sample stock level
monitor_stock(6)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(average_rainfall, required_level):
    # Comparing current rainfall against what the crops actually need
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Outputting the findings
    print(f"Average Rainfall: {average_rainfall} Rainfall Status: {status}")

# Testing with 72mm rainfall where requirement is likely lower (e.g., 50)
check_rainfall(72, 50)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    # Using a set to find unique names. If the list length differs from the set length, we have duplicates.
    if len(usernames) != len(set(usernames)):
        status = "Yes"
        # Logic: Flag accounts
    else:
        status = "No"
        # Logic: Accounts are unique

    print(f"Duplicate Accounts Found: {status}")

# Example list with a duplicate 'user1'
user_list = ["user1", "user2", "user3", "user1"]
detect_duplicate_accounts(user_list)
```

Problem 7: System Monitoring – Application Health Checker

```python
def check_app_health(error_count):
    # determining system health based on how many errors we found
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count} System Status: {status}")

# Simulating a system with 7 errors
check_app_health(7)
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    daily_limit = 50000

    # Checking if the transaction goes over the daily allowed limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount} Transaction Status: {status}")

# Trying to transfer 60,000 which is over the limit
validate_transaction(60000)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    total_bill = 0

    # Calculating bill based on slabs (Tiered pricing)
    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        # First 100 units at rate 3, remaining at rate 5
        total_bill = (100 * 3) + ((units - 100) * 5)
    else:
        # First 100 at 3, next 100 at 5, rest at 7
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Categorizing the usage based on the total bill amount
    if total_bill < 500:
        usage_level = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"

    # NOTE: The expected output format wasn't explicitly shown in the screenshot
    # for this specific problem, but I'll follow the pattern of the previous ones.
    print(f"Units Consumed: {units} Total Bill: ₹{total_bill} Usage Level: {usage_level}")

# Testing with 250 units
# Calculation: (100*3) + (100*5) + (50*7) = 300 + 500 + 350 = 1150 (Moderate)
calculate_electricity_bill(250)
```
