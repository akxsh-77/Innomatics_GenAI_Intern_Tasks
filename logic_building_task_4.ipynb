Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    # Using while loop for retry as per hint
    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))

            if amount < 50:
                print("Amount must be at least ₹50. Please try again.")
                continue

            if amount in valid_plans:
                print(f"Recharge of ₹{amount} successful!")
                break # Exit the loop if successful
            else:
                print("Invalid plan. Available plans are 199, 299, 399, 599. Try again.")

        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Run the function to test
validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(inventory_data):
    # loop through the dictionary
    for product, stock in inventory_data.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# creating a dummy dictionary to test the function
stocks = {
    "Laptops": 12,
    "Mouses": 45,
    "Keyboards": 10,
    "Monitors": 25
}

check_inventory(stocks)
```

Problem 3: Student Result Processing System

```python
def process_result(marks_list):
    total = 0

    # Using a loop to calculate total as per the hint
    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Test the code
student_marks = [45, 60, 55, 40, 70]
result = process_result(student_marks)
print(f"The student's result is: {result}")
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def estimate_cab_fare():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            peak_hour = input("Is it peak hour right now? (yes/no): ").lower()

            base_fare = 50
            per_km_charge = 12
            total_fare = base_fare + (distance * per_km_charge)

            # add 25% if it's peak hour
            if peak_hour == 'yes':
                total_fare = total_fare + (total_fare * 0.25)

            print(f"Estimated Fare: ₹{total_fare}")

            retry = input("Do you want to check another fare? (yes/no): ").lower()
            if retry != 'yes':
                print("Thank you for using the estimator!")
                break

        except ValueError:
            print("Please enter a valid number for distance.")

# Run the function to test
estimate_cab_fare()
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    present_days = 0

    # loop to count present days
    for status in attendance_list:
        if status == "P":
            present_days += 1

    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Test the function
monthly_attendance = ["P", "P", "A", "P", "P", "A", "P", "P", "P", "P"]
print("Status:", check_eligibility(monthly_attendance))
```

Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."

    has_digit = False
    has_special = False
    allowed_specials = "@#$"

    # looping over the string to check each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in allowed_specials:
            has_special = True

    # validating conditions
    if has_digit and has_special:
        return "Strong Password"
    elif not has_digit:
        return "Weak: Password must contain at least one number."
    else:
        return "Weak: Password must contain at least one special character (@#$)."

# Test the code
test_pass = input("Enter a password to check: ")
print(check_password_strength(test_pass))
```
